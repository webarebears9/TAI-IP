```python
city = stats['ship-city'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=city.index, y=city.values)
plt.title('Count of Customers in each city')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
